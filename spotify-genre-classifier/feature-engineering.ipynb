```python

def range_outliers(genre, df):
    tempo = songs.loc[songs.genre == genre, 'tempo']
    m = np.mean(tempo)
    sd = np.std(tempo)
    median = np.median(tempo)
    upper_bound = m + (sd*3)
    lower_bound = m - (sd*3)
    return[lower_bound, upper_bound]
```

It looks like some songs have tempo < 100, which is unusal for a few of the genres. Let's investigate further.

```python
fig = plt.figure(figsize=(12,8))
plt.title('Song Tempo')
plt.ylim([0,20])

some_genres = [ "Emo", "Rap", "Trap Metal", "Underground Rap", "RnB", "Hiphop", "Pop", "Dark Trap"]
for genre in some_genres:
    graph_tempo(genre, songs)
# Rnb, Hiphop, Pop, Dark Trap, Underground Rap, Trap Metal, Emo need tempo adjustment
```

Some of the Rnb, Hiphop, Pop, Dark Trap, Underground Rap, Trap Metal, and Emo tracks appear to have their tempos recorded in half time as opposed to full time.

```python
to_adjust = ["RnB", "Hiphop", "Pop", "Dark Trap", "Underground Rap", "Trap Metal", "Emo"]
def chng2full_time(data, genre):
    tempo = data.loc[data.genre == genre, 'tempo']
    threshold = max(tempo)/2
    tempo[tempo < threshold] = tempo * 2
    return tempo
```

```python
for genre in to_adjust:
    songs.loc[songs.genre == genre, 'tempo'] = chng2full_time(songs, genre)
```

To account for user error in the playlists, I will filter out songs with tempo outliers for each of the genres.

```python
def filter_tempo(data, genre, max_thresh, min_thresh):
    tempo = songs.loc[songs.genre == genre, 'tempo']
    
    above_threshold = tempo[tempo > max_thresh].index.tolist()
    below_threshold = tempo[tempo < min_thresh].index.tolist()
    indexNames = above_threshold + below_threshold
    
    data.drop(indexNames, inplace=True)
```

```python
filtered = songs.copy()
for genre in genres:
    bounds = range_outliers(genre, songs)
    filter_tempo(filtered, genre, bounds[1], bounds[0])
```
