```python
from urllib.request import urlopen
from bs4 import BeautifulSoup
html = urlopen('https://en.wikipedia.org/wiki/Gun_laws_in_the_United_States_by_state')
bs = BeautifulSoup(html, "html.parser")
```

```python
print(bs.prettify())
```

```python
tables = bs.find_all("table", class_="wikitable")[:51]
len(tables)
```

```python
headers = bs.find_all("span", class_="mw-headline")[1:52]
len(headers)
```

```python
headers[0].find(text=True)
```

```python
# print(tables[0])
law = []
long_guns = []
handguns = []
statutes = []
notes = []
states = []
    
for i in range(len(tables)):
    for row in tables[i].findAll('tr'):
        cells=row.findAll('td')
        if len(cells)==5:
            law.append(cells[0].find(text=True))
            long_guns.append(cells[1].find(text=True))
            handguns.append(cells[2].find(text=True))
            statutes.append(cells[3].find(text=True))
            notes.append(cells[4].find(text=True))
            states.append(headers[i].find(text=True))
        
```

```python
len(law)
```

```python
len(statutes)
```

```python
len(states)
```

```python
import pandas as pd
wiki_df=pd.DataFrame(states,columns=['State'])
wiki_df['Laws']=law
wiki_df['Long_guns']=long_guns
wiki_df['Handguns']=handguns
wiki_df['Statutes']=statutes
wiki_df['Notes']=notes
```
