```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt

nics = pd.read_csv('nics.csv')
census = pd.read_csv('nst-est2019-popchg2010_2019.csv')
```

```python
condensed_nics = nics.groupby(['year', 'state']).sum()
```

```python
years = np.unique(nics['year'])
states = np.unique(nics['state'])
```

```python
pop_years = range(2010,2020)
```

```python
pop_years
```

```python
#create empty population estimate column
condensed_nics['pop_est'] =""
condensed_nics.loc[2000,'Alabama']
```

```python
# condensed_nics.xs((2010, 'Alabama'))['pop_est'] = 20
condensed_nics.loc[2010, 'Alabama'] = [221911, 2]
```

```python
#fill condensed_nics df with sum of nics queries per year and population estimate values
for year in pop_years:
    temp = 'POPESTIMATE' + str(year)
    for state in states:
        tot = condensed_nics.loc[year, state][0]
        pop = census[census.NAME == state][temp].values.flatten()
        if pop:
            print(pop[0])
            condensed_nics.loc[year, state] = [tot, pop[0]]
```
