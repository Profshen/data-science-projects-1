```python
#Construct column names for combined csv
#deaths_t = total deaths, rates_t = total deaths per 100k, etc.
letters = ['t', 's', 'h', 'l']
terms = []
for i in letters:
    temp = 'deaths_' + i
    terms.append(temp)
    temp = 'rate_' + i
    terms.append(temp)
column_names = ["State", "Year", "Population"]
terms
for i in terms:
    column_names.append(i)
column_names
```

```python
combined = pd.DataFrame(columns = column_names)
combined
states_col = []
years_col = []
pop_col = []
deaths_col_t = []
rates_col_t = []
deaths_col_s = []
rates_col_s = []
deaths_col_h = []
rates_col_h = []
deaths_col_l = []
rates_col_l = []
cols = [[deaths_col_t, rates_col_t], [deaths_col_h, rates_col_h], [deaths_col_s, rates_col_s], [deaths_col_l, rates_col_l]]
```

```python
#construct arrays for each column of csv
for state in states:
    for year in years:
        total_row = total[total.State == state][total.Year == year]
        homicide_row = homicide[homicide.State == state][homicide.Year == year]
        suicide_row = suicide[suicide.State == state][suicide.Year == year]
        legal_row = legal[legal.State == state][legal.Year == year]
        rows = [total_row, homicide_row, suicide_row, legal_row]
        if total_row.shape[0]:
            states_col.append(state)
            years_col.append(year)
            pop_col.append(total_row.iloc[0]['Population'])
            for i in range(4):
                if rows[i].shape[0]:
                    temp = rows[i].iloc[0]['Deaths']
                    cols[i][0].append(temp)
                    cols[i][1].append(100000*rows[i].iloc[0]['Deaths']/rows[i].iloc[0]['Population'])
                else:
                    cols[i][0].append(0)
                    cols[i][1].append(0)
#         if total_row.shape[0]:
#             pop_col.append(total_row.iloc[0]['Population'])
#             deaths_col_t.append(total_row.iloc[0]['Deaths'])
#             rates_col_t.append(100000*total_row.iloc[0]['Deaths']/total_row.iloc[0]['Population'])
#         else:
#             rates_col_t.append(0)
#             deaths_col_t.append(0)
#         if suicide_row.shape[0]:           
#             deaths_col_s.append(suicide_row.iloc[0]['Deaths'])
#             rates_col_s.append(100000*suicide_row.iloc[0]['Deaths']/suicide_row.iloc[0]['Population'])
#         else:
#             deaths_col_s.append(0)
#             rates_col_s.append(0)
#         if homicide_row.shape[0]:
#             deaths_col_h.append(homicide_row.iloc[0]['Deaths'])
#             rates_col_h.append(100000*homicide_row.iloc[0]['Deaths']/homicide_row.iloc[0]['Population'])
#         else:
#             deaths_col_h.append(0)
#             rates_col_h.append(0)
#         if legal_row.shape[0]:
#             deaths_col_l.append(legal_row.iloc[0]['Deaths'])
#             rates_col_l.append(100000*legal_row.iloc[0]['Deaths']/legal_row.iloc[0]['Population'])
#         else:
#             deaths_col_l.append(0)
#             rates_col_l.append(0)
        
        
```
